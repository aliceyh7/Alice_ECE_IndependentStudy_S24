```python
# Given data from the user
data = {
    'Marker Type': ['ArUco Marker', 'ArUco Marker', 'ArUco Marker',
                    'Colored Marker', 'Colored Marker', 'Colored Marker',
                    'Unity Logo Marker', 'Unity Logo Marker', 'Unity Logo Marker'],
    'Data Type': ['Simulated Dataset with Unity', 'Real-World Dataset', 'Combined',
                  'Simulated Dataset with Unity', 'Real-World Dataset', 'Combined',
                  'Simulated Dataset with Unity', 'Real-World Dataset', 'Combined'],
    'Precision': [0.8548758234, 0.8865979381, 0.8977597817,
                  0.3870129870, 0.3571623847, 0.4,
                  0.7248894038, 0.7, 0.75],
    'Recall': [0.9973544937, 0.9608938547, 0.9602837154,
               0.4257142857, 0.4071623998, 0.45,
               0.7492047264, 0.76, 0.77],
    'F1 Score': [0.9206349206, 0.9222501027, 0.8673205756,
                 0.4054421768, 0.3762384765, 0.42,
                 0.7124084262, 0.73, 0.76],
    'mAP Score': [0.901, 0.883, 0.892,
                  0.413, 0.402, 0.43,
                  0.780, 0.79, 0.80]
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Define the bar width and positions
bar_width = 0.2
index = pd.Index([1, 2, 3])
r1 = index - bar_width
r2 = index
r3 = index + bar_width

# Plotting function with Arial font if available
def plot_metric(df, metric, color, ax):
    # Filter the DataFrame for the simulated dataset
    simulated_data = df[df['Data Type'] == 'Simulated Dataset with Unity'][metric]
    # Filter the DataFrame for the real-world dataset
    real_world_data = df[df['Data Type'] == 'Real-World Dataset'][metric]
    # Filter the DataFrame for the combined dataset
    combined_data = df[df['Data Type'] == 'Combined'][metric]

    # Plot each dataset type for the metric
    ax.bar(r1, simulated_data, width=bar_width, label='Simulated', color=color[0])
    ax.bar(r2, real_world_data, width=bar_width, label='Real-World', color=color[1])
    ax.bar(r3, combined_data, width=bar_width, label='Combined', color=color[2])

    # Add titles and labels with increased font size for titles
    ax.set_xlabel('Marker Type', fontsize=12)
    ax.set_ylabel(metric, fontsize=12)
    ax.set_title(f'{metric} for Each Marker Type', fontsize=18, fontname='Arial')
    ax.set_xticks(index)
    ax.set_xticklabels(['ArUco', 'Colored', 'Unity Logo'])
    ax.legend()

# Creating subplots for each metric
fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # Adjust the size as needed

# Plot for each metric
plot_metric(df, 'Precision', ['#4285f4', '#ea4335', '#fbbc05'], axs[0, 0])
plot_metric(df, 'Recall', ['#4285f4', '#ea4335', '#fbbc05'], axs[0, 1])
plot_metric(df, 'F1 Score', ['#4285f4', '#ea4335', '#fbbc05'], axs[1, 0])
plot_metric(df, 'mAP Score', ['#4285f4', '#ea4335', '#fbbc05'], axs[1, 1])

# Layout adjustment
plt.tight_layout()
plt.show()
```
