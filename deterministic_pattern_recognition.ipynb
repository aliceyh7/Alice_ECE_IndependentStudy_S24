```python
import pandas as pd
import numpy as np
import ast
import os
import matplotlib.pyplot as plt

# Assuming the data directory structure is as shown before
data_directory = './'  # Set to your data directory

# Define a function to read marker ids and corners from the text files
def read_marker_data(trial_num):
    corners_file = f'trial{trial_num}_marker_corners.txt'
    ids_file = f'trial{trial_num}_marker_ids_and_corners.txt'

    # Read corners data
    corners_df = pd.read_csv(os.path.join(data_directory, corners_file), header=None, names=['timestamp', 'x', 'y'])

    # Convert 'timestamp' to string to ensure consistent data types
    corners_df['timestamp'] = corners_df['timestamp'].astype(str)

    # Read ids and corners data
    ids_data = []
    with open(os.path.join(data_directory, ids_file), 'r') as file:
        for line in file:
            if 'Corners' in line:
                timestamp, rest = line.strip().split(' ID: ')
                marker_id, corners_str = rest.split(' Corners: ')
                corners_list = ast.literal_eval(corners_str)
                x_avg = np.mean([corner[0] for corner in corners_list])
                y_avg = np.mean([corner[1] for corner in corners_list])
                ids_data.append([timestamp, int(marker_id), x_avg, y_avg])

    ids_df = pd.DataFrame(ids_data, columns=['timestamp', 'marker_id', 'x_avg', 'y_avg'])

    # Convert 'timestamp' to string to ensure consistent data types
    ids_df['timestamp'] = ids_df['timestamp'].astype(str)

    # Merge both dataframes on timestamp
    merged_df = corners_df.merge(ids_df, on='timestamp', how='inner')
    return merged_df

# Load data for each trial
trial_data = {}
for i in range(1, 10):  # We have 9 trials
    trial_data[i] = read_marker_data(i)

# Placeholder code for comparison and visualization
# Here you will need to compare the merged data for each trial with the baseline
# and visualize the result. The implementation of this part is left as an exercise
# because it will depend on the specific analysis and comparison you want to perform.

# ... (previous code) ...

# Assuming trial 1 is the baseline
baseline_data = trial_data[1]

# Initialize a dictionary to hold comparison metrics
comparison_metrics = {}

# Define the metric calculation - in this case, we'll use mean error
for trial_num, data in trial_data.items():
    comparison_metrics[trial_num] = {
        'mean_x': data['x_avg'].mean(),
        'mean_y': data['y_avg'].mean()
    }

# Create DataFrames from the metrics for easier manipulation
baseline_metrics = pd.DataFrame([comparison_metrics[1]])
trial_metrics = pd.DataFrame.from_dict(comparison_metrics, orient='index')

# Calculate differences from baseline for trials
trial_metrics['delta_mean_x'] = trial_metrics['mean_x'] - baseline_metrics['mean_x'].values[0]
trial_metrics['delta_mean_y'] = trial_metrics['mean_y'] - baseline_metrics['mean_y'].values[0]

# Visualization: Plot the differences from baseline for each trial
plt.figure(figsize=(14, 7))

# Plot for delta in x
plt.subplot(1, 2, 1)
plt.bar(trial_metrics.index, trial_metrics['delta_mean_x'], color='skyblue')
plt.xlabel('Trial Number')
plt.ylabel('Delta Mean X Position')
plt.title('Delta Mean X Position from Baseline Across Trials')

# Plot for delta in y
plt.subplot(1, 2, 2)
plt.bar(trial_metrics.index, trial_metrics['delta_mean_y'], color='lightgreen')
plt.xlabel('Trial Number')
plt.ylabel('Delta Mean Y Position')
plt.title('Delta Mean Y Position from Baseline Across Trials')

plt.tight_layout()
plt.show()
```

```python
# Function to calculate detection accuracy
def calculate_detection_accuracy(trial_data, total_images):
    # Count the number of unique images where markers were detected
    detected_images = trial_data['timestamp'].nunique()
    # Calculate the accuracy
    accuracy = (detected_images / total_images) * 100
    return accuracy

# Total number of images per trial (based on your table)
total_images = 1382

# Dictionary to hold accuracy for each trial
accuracy_per_trial = {}

# Calculate accuracy for each trial
for trial_num in range(1, 10):
    trial_df = trial_data[trial_num]  # Data for the current trial
    accuracy_per_trial[trial_num] = calculate_detection_accuracy(trial_df, total_images)

# Output the accuracy results
for trial_num, accuracy in accuracy_per_trial.items():
    print(f"Trial {trial_num}: {accuracy:.2f}% accuracy")

# Visualization: Plot the accuracy for each trial

plt.figure(figsize=(12, 6))  # Wider and shorter figure size
plt.bar(accuracy_per_trial.keys(), accuracy_per_trial.values(), color='orange')
plt.xlabel('Trial Number')
plt.ylabel('Detection Accuracy (%)')
plt.title('Detection Accuracy Across Trials', fontsize=20)  # Bigger title text
plt.ylim(0, 100)  # Assuming accuracy is a percentage
plt.tight_layout()  # Ensure everything fits without overlapping
plt.show()
```

**Conclusions from Data (Results)**

1. Variation in Detection Accuracy:
(a) The highest detection accuracy was achieved in Trial 4, where the focal length was increased, suggesting that increasing the focal length might help in improving detection accuracy.
(b) Trial 2, which had a larger marker size, also showed a notable increase in detection accuracy compared to the baseline.
(c) However, reducing the marker size (Trial 3) or reducing the focal length (Trial 5) led to lower detection accuracies, indicating that smaller markers or focal lengths hinder accurate detection.


2. Impact of Camera Position and Angle:

Camera position and angle were significantly altered in Trials 8 and 9. These adjustments did not seem to have a positive effect on detection accurac. This is suggesting that the initial camera position and angle were close to optimal for this particular setup.

3. Sensor Size Influence:

In Trial 6, the sensor size was increased, there was a slight decrease in detection accuracy compared to the baseline. However, in Trial 7, with a reduced sensor size, detection accuracy dropped even more. It seems that there is a specific sensor size range where detection accuracy is maximized.


4. Consistency Across Trials:

The first image shows the marker corner positions and the average positions across trials. There is a visible dispersion of marker positions, particularly in Trial 6, which could be due to the change in sensor size impacting the field of view and thus the perceived marker position.


Overall, the data suggests that Larger markers and longer focal lengths improve accuracy, while smaller markers and shorter focal lengths reduce it.

Also, the initial camera position and angle seem to be effective, as changes did not improve accuracy.

Sensor size has variable relationship with detection accuracy, and there appears to be an optimal range that maximizes accuracy.
